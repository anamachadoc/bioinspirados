```python
distance = {}
row_index = 1  # Começamos no 1

with open('input/LAU15/lau15_dist.txt', 'r') as file:
    for line in file:
        line = line.strip()
        if not line or line.startswith('#'):
            continue

        numbers = list(map(int, line.split()))
        distance[row_index] = {}
        for col_index, num in enumerate(numbers, start=1):  # colunas também começam em 1
            distance[row_index][col_index] = num

        row_index += 1
```

```python
# Abrindo o arquivo
with open('input/LAU15/lau15_xy.txt', 'r') as f:
    lines = f.readlines()

# Inicializando o dicionário
xy = {}

# Processando as linhas do arquivo
for i, line in enumerate(lines, start=1):
    # Ignora linhas vazias ou mal formatadas
    if line.strip():
        # Split para separar as coordenadas
        coords = line.split()
        # Convertendo as coordenadas para float
        x = float(coords[0])
        y = float(coords[1])
        # Adiciona no dicionário com a chave como o índice da cidade
        xy[i] = (x, y)

# Exibindo o dicionário
print(xy)
```

```python
import random
```

```python
pai1 = [0,1,5,3,4,2]
pai2= [3,5,4,2,1,0]
```

```python
p1, p2 = sorted(random.sample(range(6), 2))
print(p1,p2)
```

```python
segment1 = pai1[p1:p2]
segment2 = pai2[p1:p2]
print(segment1, segment2)
```

```python
def preencher(base, fixo):
    return [x for x in base if x not in fixo]
rest1 = preencher(pai2, segment1)
rest2 = preencher(pai1, segment2)
```

```python
print(rest1, rest2)
```

```python
child1 = rest1[:p1] + segment1 + rest1[p1:]
child2 = rest2[:p1] + segment2 + rest2[p1:]
```
